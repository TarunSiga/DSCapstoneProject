```python
# Preprocessing function
def preprocess(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalnum()]
    # Convert to lowercase
    filtered_text = ' '.join(filtered_tokens).lower()
    return filtered_text

```

```python
# Calculate similarity between two texts using cosine similarity
def calculate_similarity(text1, text2):
    # Tokenize and preprocess the texts
    text1 = preprocess(text1)
    text2 = preprocess(text2)
    # Vectorize the texts
    vectorizer = CountVectorizer().fit_transform([text1, text2])
    vectors = vectorizer.toarray()
    # Calculate cosine similarity
    similarity = cosine_similarity(vectors[0].reshape(1,-1), vectors[1].reshape(1,-1))
    return similarity[0][0]
```

```python
# Preprocessing function
def preprocess(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalnum()]
    # Convert to lowercase
    filtered_text = ' '.join(filtered_tokens).lower()
    return filtered_text

```

```python
# Calculate similarity between two texts using cosine similarity
def calculate_similarity(text1, text2):
    # Tokenize and preprocess the texts
    text1 = preprocess(text1)
    text2 = preprocess(text2)
    # Vectorize the texts
    vectorizer = CountVectorizer().fit_transform([text1, text2])
    vectors = vectorizer.toarray()
    # Calculate cosine similarity
    similarity = cosine_similarity(vectors[0].reshape(1,-1), vectors[1].reshape(1,-1))
    return similarity[0][0]
```
